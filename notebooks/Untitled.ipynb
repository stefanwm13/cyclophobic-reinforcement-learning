```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set_theme(style="darkgrid")
sns.set_style("white")
```

```python
doorkey_11x11_1 = pd.read_csv('results/DoorKey-11x11_new.csv', sep=" ")  
#doorkey_11x11_2 = pd.read_csv('results/DoorKey-Rand-11x11_2.csv', sep=" ")  
#doorkey_11x11_3 = pd.read_csv('results/DoorKey-Rand-11x11_3.csv', sep=" ")  


doorkey_8x8_1 = pd.read_csv('results/DoorKey-Rand-8x8.csv', sep=" ")  
doorkey_8x8_2 = pd.read_csv('results/DoorKey-Rand-8x8_2.csv', sep=" ")  
doorkey_8x8_3 = pd.read_csv('results/DoorKey-Rand-8x8_3.csv', sep=" ")  


unlockpickup_8x8 = pd.read_csv('results/UnlockPickup-8x8.csv', sep=" ")  
unlockpickup_8x8['reward'] = unlockpickup_8x8['reward'].rolling(100).mean()


doorkey_11x11_1 = pd.DataFrame({'steps':doorkey_11x11_1['steps'], 'reward':doorkey_11x11_1['reward'].rolling(100).mean()})
#doorkey_11x11_2 = pd.DataFrame({'steps':doorkey_11x11_1['steps'], 'reward':doorkey_11x11_2['reward'].rolling(100).mean()})
#doorkey_11x11_3 = pd.DataFrame({'steps':doorkey_11x11_1['steps'], 'reward':doorkey_11x11_3['reward'].rolling(100).mean()})


doorkey_8x8_1_pd = pd.DataFrame({'steps':doorkey_8x8_1['steps'], 'reward':doorkey_8x8_1['reward'].rolling(100).mean()})
doorkey_8x8_2_pd = pd.DataFrame({'steps':doorkey_8x8_1['steps'], 'reward':doorkey_8x8_2['reward'].rolling(100).mean()})
doorkey_8x8_3_pd = pd.DataFrame({'steps':doorkey_8x8_1['steps'], 'reward':doorkey_8x8_3['reward'].rolling(100).mean()})

```

```python
doorkey_8x8 = pd.concat([doorkey_8x8_1_pd.assign(region='North'), doorkey_8x8_2_pd.assign(region='North'), doorkey_8x8_3_pd.assign(region='North')])
doorkey_8x8['C']=np.arange(len(doorkey_8x8))
doorkey_8x8

doorkey_11x11 = pd.concat([doorkey_11x11_1.assign(region='North')])#, doorkey_11x11_2.assign(region='North'), doorkey_11x11_3.assign(region='North')])
doorkey_11x11['C']=np.arange(len(doorkey_11x11))
doorkey_11x11
```

```python
sns.lineplot(data=doorkey_11x11, x="steps", y="reward", ci=95)

```
