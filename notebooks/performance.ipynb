```python
import pandas as pd

# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline

# Seaborn for plotting and styling
import seaborn as sns

import numpy as np
import pickle

from minigrid_env import *
from gym import wrappers

from IPython import display
%matplotlib inline

from IPython.display import Video

from IPython.display import Image
```

# S25

```python
df1 = pd.read_csv("csv_results/S25/S25_noexp_0.csv", sep=';')
df2 = pd.read_csv("csv_results/S25/S25_noexp_1.csv", sep=';')
df3 = pd.read_csv("csv_results/S25/S25_noexp_2.csv", sep=';')


df1_nls = pd.read_csv("csv_results/S25/S25_normal_0.csv", sep=';')


df1_n = pd.read_csv("csv_results/S25/S25_normal_0.csv", sep=';')
```

```python
df1_roll = pd.DataFrame({'steps':df3['steps'], 'reward':df1['reward'].rolling(10).mean()})
df2_roll = pd.DataFrame({'steps':df3['steps'], 'reward':df2['reward'].rolling(10).mean()})
df3_roll = pd.DataFrame({'steps':df3['steps'], 'reward':df3['reward'].rolling(10).mean()})

df_total_1 = pd.concat([df1_roll, df2_roll, df3_roll])

df_end_1 = pd.DataFrame({'steps':df_total_1['steps'], 'reward':df_total_1['reward']})
df_end_1['cycle'] = "Cycle"

df_total_1_nls = pd.DataFrame({'steps':df1_nls['steps'], 'reward':df1_nls['reward'].rolling(1).mean()})

df_end_1_nls = pd.DataFrame({'steps':df_total_1_nls['steps'], 'reward':df_total_1_nls['reward']})
df_end_1_nls['cycle'] = "Cycle w/o normalization"

df_total_1_n = pd.DataFrame({'steps':df1_n['steps'], 'reward':df1_n['reward'].rolling(1).mean()})

df_end_1_n = pd.DataFrame({'steps':df_total_1_n['steps'], 'reward':df_total_1_n['reward']})
df_end_1_n['cycle'] = "Q-Learning"


df_final = pd.concat([df_end_1, df_end_1_nls, df_end_1_n])
```

```python
sns.lineplot(data=df_final, x="steps", y="reward", hue="cycle", ci=95)
```
